```
import random
import string
import os
from concurrent.futures import ThreadPoolExecutor
```

# **Data Generator**

```
def gen_rand_var(var_type,length):
  if var_type is int:
    range_start = 10**(length-1)
    range_end = (10**length)-1
    var = random.randint(range_start, range_end)
  elif var_type is float:
    range_start = 10**(length-1)+0.1
    range_end = (10**length)-1 + 0.9
    var = round(random.uniform(range_start, range_end),3)
  else:
    var = ''.join(random.choice(string.ascii_uppercase + string.ascii_lowercase + string.digits) for _ in range(length))
  return var
# print(gen_rand_var(str,16))
```

```
max_length = 10
file_size = 0
file_path = '/content/mFile.txt'
#open file
with open(file_path,'a') as mFile:
  while file_size < 1e+7:
    types = [str,float,int]
    t = random.choice(types)
    length = random.randint(1,max_length)
    v = gen_rand_var(t,length)
    if t is str: # adding random spaces in alphanumeric
      if not v.isalpha():
        spaces = random.randint(1,9)
        v = (' '*spaces) + v + (' '*spaces)
    mFile.write(f'{v}, ')
    file_size = os.stat(file_path).st_size
  mFile.write('123') # making sure file doesnt end with a space
  mFile.close()
```

# **Data Reader**

```
with open("/content/mFile.txt", "r") as mFile:
  text = mFile.read()
  text = text.replace(' ','')
  items = text.split(',')

def fetch(obj):
  if obj.isnumeric():
    res = f'{obj} - integer'
  elif obj.isalpha():
    res = f'{obj} - alphabetical strings'
  elif obj.isalnum():
    res = f'{obj} - alphanumeric'
  else:
    res = f'{obj} - real number'
  return res

pool = ThreadPoolExecutor(max_workers=10)

for j in pool.map(fetch, items):
  print(j)
```
